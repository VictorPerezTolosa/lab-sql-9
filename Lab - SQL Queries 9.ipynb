### Create a Python connection with SQL database and retrieve the results of the last two queries (also mentioned below) as dataframes:

* Check the number of rentals for each customer for May

* Check the number of rentals for each customer for June

```python
# Import libraries and dependencies

import mysql.connector
import pandas as pd

from mysql.connector import connect
from getpass import getpass
```

```python
# Variable for the SQL database

db = connect(
        host= "localhost",
        user = "root",
        passwd = getpass("Enter password: "),
        database= "sakila"
   )
```

```python
cursor = db.cursor() # To execute SQL STATEMENTS on python

cursor.execute("SHOW DATABASES")
```

```python
databases= cursor.fetchall()
print(databases)
```

* Check the number of rentals for each customer for May

```python
customers_may_query = "select customer_id, count(*) as 'number of rentals' from rentals_may group by customer_id order by customer_id"

cursor.execute(customers_may_query)
```

```python
may_df = cursor.fetchall()
```

```python
may_customers = pd.DataFrame(may_df)
may_customers
```

* Check the number of rentals for each customer for June

```python
customers_june_query = "select customer_id, count(*) as 'number of rentals' from rentals_june group by customer_id order by customer_id;"

cursor.execute(customers_june_query)
```

```python
june_df = cursor.fetchall()
```

```python
june_customers = pd.DataFrame(june_df)
june_customers
```

* Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

```python
merged_customers = may_customers.merge(june_customers, how = 'outer', on = 0, sort = True)
merged_customers
```

```python
# Changing the header's names

merged_customers.rename(columns={0: 'customer_id', '1_x': 'may', '1_y': 'june'}, inplace=True)
merged_customers
```

```python
merged_customers = merged_customers.fillna(0) # Filling the NaN with 0
merged_customers
```

```python
# creating the function, I will create a new column where I will put all the result of each comparison in

def month_comparison(x):
    import pandas as pd
    import numpy as np
    x["comparison"] = np.where((x.iloc[:,1] < x.iloc[:,2]), "more", "same or less")
    display(x)
```

```python
month_comparison(merged_customers)
```
